# Joining final data sets


```python
import os
import pandas as pd
from functools import reduce
```

```python
# Loading the final datasets data come as a result of the jupiter notebooks contained in  
FRSquest = pd.read_csv('E:\HumanA\Data\Data_Tracy\Questionnaires\Adjusted\FRSsubs.csv')
FRSquest.drop('Unnamed: 0', axis=1, inplace=True)
FRSquest.head()
```

```python
affiliative = pd.read_csv('E:\HumanA\Data\Data_Tracy\Questionnaires\Adjusted/AffilliativeTend.csv')
affiliative = affiliative[['ID', 'AffiliativeQ_average']]
affiliative.head(18)
```

```python
context = pd.read_csv('E:\HumanA\Data\Data_Tracy\Questionnaires\Adjusted/ContextQuest.csv')
context = context[['ID', 'item3','ContextQ_average']]
context.head()
```

```python
avatar_sum = pd.read_csv('E:\HumanA\Data\Data_Tracy\Questionnaires\Adjusted\AgentQuest.csv')
avatar_sum.head(21)
```

```python
# TODO: add pointing summary for Building and Agent Task & rename RT to RT_building/ RT_agent, as well as AbsolutError
building_pointing_sum= pd.read_csv('E:\HumanA\Data\Data_Tracy\BuildingTask\BuildingPointingSummary.csv')
building_pointing_sum.drop('Unnamed: 0', axis=1, inplace=True)
building_pointing_sum.head()
```

```python
building_pointing_sum=building_pointing_sum.set_index(['ID']).add_prefix('Building_').reset_index()
```

```python
building_pointing_sum.info()
```

```python
# TODO: add pointing summary for Building and Agent Task & rename RT to RT_building/ RT_agent, as well as AbsolutError
agent_pointing_sum= pd.read_csv('E:\HumanA\Data\Data_Tracy\AgentTask\AgentPointingSummary.csv')
agent_pointing_sum.drop('Unnamed: 0', axis=1, inplace=True)
agent_pointing_sum.head()
```

```python
agent_pointing_sum=agent_pointing_sum.set_index(['ID']).add_prefix('Agent_').reset_index()
```

```python
agent_pointing_sum.info()
```

```python
#data_frames = [FRSquest, affiliative, context, avatar_sum, building_pointing_sum, agent_pointing_sum]
data_frames = [FRSquest, affiliative, context, avatar_sum, building_pointing_sum]
humanA_Behavioural_Building = reduce(lambda  left,right: pd.merge(left,right,on=['ID'],
                                                how='outer'), data_frames)
```

```python
data_frames = [FRSquest, affiliative, context, avatar_sum, agent_pointing_sum]
humanA_Behavioural_Agent = reduce(lambda  left,right: pd.merge(left,right,on=['ID'],
                                                how='outer'), data_frames)
```

```python
humanA_Behavioural_Agent.info()
```

```python
humanA_Behavioural_Building.to_csv('E:\HumanA\Data\Data_Tracy\Resulting_CSVs\humanA_Behavioural_Building.csv')
humanA_Behavioural_Agent.to_csv('E:\HumanA\Data\Data_Tracy\Resulting_CSVs\humanA_Behavioural_Agent.csv')
```
