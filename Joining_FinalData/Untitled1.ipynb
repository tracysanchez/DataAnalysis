```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data = pd.DataFrame({'Previous_response': [ "Right","Right","Left","Left"], 'Predicted_Probability': [0.77, 0.72, 0.72,0.77], "Previous_Trial_Identity": ['Own', "Partner",'Own',"Partner" ], "Standard_Error": [0.1,0.1,0.1,0.1] })
Data_Left= Data[Data.Previous_response == "Left"]
Data_Right= Data[Data.Previous_response == "Right"]
```

```python
plt.style.use('default')
plt.rcParams["font.family"] = "Arial"

#fig, axes = plt.subplots(1, 1, sharey=True,  figsize=(8, 6))




plt.errorbar( x=Data_Left["Previous_Trial_Identity"],y=Data_Left["Predicted_Probability"].values,
             yerr=Data_Left.Standard_Error, fmt='o',  color='#9b2226', label= "Left", markersize=7, capthick=20)
plt.errorbar( x=Data_Right["Previous_Trial_Identity"],y=Data_Right["Predicted_Probability"].values,
             yerr=Data_Right.Standard_Error, fmt='o',  color='#577590', label= "Right", markersize=7, capthick=20)
plt.ylabel("Predicted probability")
plt.title('TITLE', fontsize=16)
plt.legend(loc='upper center')
plt.xlim((-0.5, 1.5))
plt.ylim((0, 1))
plt.grid(True, color = "grey", linewidth = "0.3", linestyle = "-.")
plt.tick_params(axis='both', labelsize=16)
```
