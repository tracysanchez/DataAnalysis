## Dealing with time 
- Create continues time 
- Delete duplicate time stamps
- Create individual csv files per subject per session

```python
import pandas as pd 
import numpy as np
```

```python
result =  pd.load_csv("/Volumes/SSD/00_Data_Processing/Pre_processed/Complete_data_Categorical.csv")
```

```python
Non_duplicate_data.drop(['Unnamed: 0', 'level_0', 'index'], axis=1, inplace=True)
```

```python
# Group by Subject and Session because the continuous time only makes sence on a session scale
all_participants_sessions = pd.DataFrame(result.groupby(["SubjectID", "Session"])["hitColliderType"].count())
all_participants_sessions.reset_index(inplace=True)
# Create tuples to later subset data to subject-session scale. We create this because subjects can have a different number of sessions, in this way we only subset for the existing pairs S-S.
L_all_participants_sessions = list(zip(all_participants_sessions.SubjectID, all_participants_sessions.Session))
```

```python
len(L_all_participants_sessions)
```

```python
temporals = []
for tuples in L_all_participants_sessions:
    subject, session = tuples
    temporal = result[(result['SubjectID'] == subject) & (result['Session'] == session)]
    temporal_c = temporal.copy(deep=True)
    temporal_c["Continuous_Time"] = np.round((temporal_c.iloc[:, 5] - temporal_c.iloc[0,5])/60, 3)
    # Delete duplicate time stamps
    temporal_c["Time_Shift"] = temporal_c.timeStampDataPointEnd.diff()
    temporal_c_ND =  temporal_c[temporal_c["Time_Shift"] > 0.0000]
    temporal_c_ND.to_csv(f"/Volumes/SSD/00_Data_Processing/Pre_processed/03_Individuals_IndividualSessions/{subject}_{session}.csv", index=True)
    temporals.append(temporal_c_ND)
    print(tuple)
```

```python
data_frames_list = pd.concat(temporals, axis=0)
```

```python
One_participant = pd.read_csv("/Volumes/SSD/00_Data_Processing/Pre_processed/03_Individuals_IndividualSessions/479_5.csv")
```

```python
def bitmask_flag_change(Data_Frame):
    '''
    '''
  change_flag = ["No_change"] # initialize flag 
  for i in range(1, len(Data_Frame)):
    if Data_Frame["combinedGazeValidityBitmask"][i] == Data_Frame["combinedGazeValidityBitmask"][i-1]: # compare each value with the previous value
      change_flag.append("No_change") 
    elif Data_Frame["combinedGazeValidityBitmask"][i] != Data_Frame["combinedGazeValidityBitmask"][i-1]:
        if Data_Frame["combinedGazeValidityBitmask"][i] == 3:
              change_flag.append("Invalid_Section_Ends")
        elif Data_Frame["combinedGazeValidityBitmask"][i] == 0:
            change_flag.append("Invalid_Section_Starts")
        else:
            change_flag.append("ERROR")
  return change_flag
```

```python
One_participant["Bitmask_flag"] = bitmask_flag_change(One_participant)
```

```python
#Non_duplicate_data.reset_index(inplace=True)
#Drop all the second hit colliders that do not comply with the criteria
indexLastValid = One_participant[One_participant["Bitmask_flag"] == "Invalid_Section_Starts"].index
indexLastInValid = One_participant[One_participant["Bitmask_flag"] == "Invalid_Section_Ends"].index
indexLower = indexLastValid - 20
indexUpper = indexLastValid
```

```python
# Since it's # Since it's possible that the invalid event occured less than 20
# rows from the beggining of the file we need to correct so that the highest possible index is the firs value in the file
for index, element in enumerate(indexLower[0:20]):
    if element < One_participant.index[0]:
        indexLower[index] = One_participant.index[0]
```

```python
Elements_to_replace = [One_participant.iloc[down:up,23].mode().iloc[0] for down, up in zip(indexLower, indexUpper)]
```

```python
len(Elements_to_replace)
```

```python
One_participant["Interpolated_collider"] = One_participant["hitObjectColliderName"]
ranges = list(zip(indexLastValid, indexLastInValid))
for i, (lower, upper) in enumerate(ranges):
    One_participant.iloc[lower:upper,36]  = Elements_to_replace[i]
```
